```python
#Ans 3
'''The try and except Block: Handling Exceptions
The try and except block in Python is used to catch and handle exceptions. 
Python executes code following the try statement as a “normal” part of the program. 
The code that follows the except statement is the program’s response to any exceptions in the preceding try clause.
'''
try:
    numerator = 10
    denominator = 0

    result = numerator/denominator

    print(result)
except:
    print("Error: Denominator cannot be 0.")

#Output: Error: Denominator cannot be 0. 
```

```python
#Ans 4
'''
Try: This block will test the excepted error to occur
Else: If there is no exception then this block will be executed
Finally: Finally block always gets executed either exception is generated or not
Raise: The raise keyword is used to raise an exception. 
You can define what kind of error to raise, and the text to print to the user.
'''

# Python code to illustrate
# working of try()
def divide(x, y):
	try:
		# Floor Division : Gives only Fractional
		# Part as Answer
		result = x // y
	except ZeroDivisionError:
		print("Sorry ! You are dividing by zero ")
	else:
		print("Yeah ! Your answer is :", result)


divide(3, 2)
divide(3, 0)


# Python code to illustrate
# working of try()
   
def divide(x, y):
    try:
        # Floor Division : Gives only Fractional
        # Part as Answer
        result = x // y
    except ZeroDivisionError:
        print("Sorry ! You are dividing by zero ")
    else:
        print("Yeah ! Your answer is :", result)
    finally:
        # this block is always executed
        # regardless of exception generation.
        print('This is always executed')

        
divide(3, 2)
divide(3, 0)

#Raise Example

s = 'apple'

try:
	num = int(s)
except ValueError:
	raise ValueError("String can't be changed into integer")
```

```python
#Ans 5
'''
Custom exceptions in Python are user-defined exceptions that can be created by the programmer to handle 
specific error situations.
They allow programmers to define their own types of exceptions based on their application requirements.

There are several reasons why we need custom exceptions:
1. Specific error handling: Custom exceptions provide more specific information about the error and help in better 
understanding and debugging of the code.
2. Modularity: They allow for better code organization and encapsulation, as related exceptions can be grouped together.
3. Customized error messages: Custom exceptions can include additional information or custom error messages that are
more meaningful for the application.

'''
 
class Error(Exception):
    """Base class for other exceptions"""
    pass
 
class zerodivision(Error):
    """Raised when the input value is zero"""
    pass
 
try:
    i_num = int(input("Enter a number: "))
    if i_num == 0:
        raise zerodivision
except zerodivision:
    print("Input value is zero, try again!")
    print()
```

```python
#Ans 6
'''
Let's say we have a program that calculates the average of a list of numbers. 
However, we want to raise a custom exception when the list is empty, as there is no average to calculate in such a case.
'''

class EmptyListError(Exception):
    pass

def calculate_average(numbers):
    if not numbers:
        raise EmptyListError("The list is empty. Cannot calculate average.")
    total = sum(numbers)
    average = total / len(numbers)
    return average

try:
    numbers = [1, 2, 3, 4, 5]
    average = calculate_average(numbers)
    print(f"The average is: {average}")
    
    empty_list = []
    average = calculate_average(empty_list)
    print(f"The average is: {average}")
except EmptyListError as e:
    print(str(e))
'''
In this example, we define a custom exception called `EmptyListError`.
In the `calculate_average` function, we check if the list of numbers is empty. 
If it is, we raise the `EmptyListError` exception with a custom error message.
When we call the `calculate_average` function with a non-empty list, the average is calculated and printed.
However, when we call it with an empty list, the custom exception is raised and caught in the `except` block,
where we print the custom error message.
The use of a custom exception here allows for a more specific and informative error message when an empty
list is encountered, making it easier to handle and identify the issue.
'''   
    
```
