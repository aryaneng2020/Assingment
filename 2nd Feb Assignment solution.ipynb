```python
Ans 1: My universal rule for choosing between a while loop and for loop
is that I use while loops
when I do not know the number of iterations ahead of time and
for loops when I do know the number of iterations ahead of time.

For loop Example:-

# input list
# Here we do know the number of iterations ahead of time; so we had use the for loop. 
inputList = [10, 20, 30, 40, 50]
print("Input list elements:")
# traversing through all elements of the list using for loop
for element in inputList:
   # printing each element of the list
   print(element)
   
   
While loop Example:-

# Here we do not know the number of iterations ahead of time; so we had use the while loop. 
# Initializing a dummy variable with 1
i = 1
# Iterate until the given condition is true
while i < 10:
   # printing the current value of the above variable
   print(i)
   # incrementing the value of i value by 1
   i += 1 
```

```python
#Ans 2:
#for Loop program 
#Sum and multiplication of natural numbers up to num


n=10
sum = 0
mul = 1 
for i in range(1, n + 1):
    sum = sum + i
    mul *= i
print("The sum of First 10 Natural digit no. from for loop method is", sum)
print("The mul of First 10 Natural digit no. from for loop method is", mul)



#While loop program
#Sum and multiplication of natural numbers up to num

num = 10
if num < 0:
   print("Enter a positive number")
else:
    sum = 0
    mul=1 
    # use while loop to iterate until zero
    while(num > 0):
        sum += num
        mul *=num
        num -= 1
    print("The sum of First 10 Natural digit no. from while loop method is", sum)
    print("The mul of First 10 Natural digit no. from while loop method is", mul)
```

```python
#Ans 4:

print("For loop Method:-")
'''For loop Method:-'''
l=[]
for i in range(1,101):
    l.append(i)

print("List of no. 1 to 100:-",l)

for i in range (1,101):
    m=[i**3 for i in l]

l1=[]
for i in m:
    if (i%4==0 or i%5==0):
        l1.append(i)
print("")
print("List of the cube of that number 1 to 100 is divisible by 4 or 5:-\n",l1)

print("")
print("While Loop Method:-")
'''While Loop Method'''
l=[]
i=1
while i<101:
    l.append(i)
    i+=1
print("List of no. 1 to 100:-",l)

while i<101:
    m=[i**3 for i in l]

l2=[]
i1 = 0
while i1 < len(m):
    e = m[i1]
    i1 += 1
    if (e%4==0 or e%5==0):
        l2.append(e)

print("")
print("List of the cube of that number 1 to 100 is divisible by 4 or 5:-\n",l2)
```
